```python
class Solution:
    def removeElement(self, nums, val):
        k = 0
        for l in nums:
            if l != val:
                nums[k] = l
                k += 1
        return k

solution = Solution()
result = solution.removeElement([3,2,2,3], 3)   
print(result)  # Output: 2
```
